```python
#toltal number of wounded people in INDIA

import csv
with open("year2017.csv","r") as file_obj:
    file_data=csv.DictReader(file_obj,skipinitialspace=True)
    list_file=list(file_data)
wounded={}
for row in list_file:
    key=row["Country"]
    if key=="India":
        key="India"
    else:
        key="other"
        
        
    value=row["Wounded"]
    if value!="":
        value=int(float(value))
    else:
        value=0
    if key in wounded:
        wounded[key]+=value
    else:
        wounded[key]=value
wounded
    
```
