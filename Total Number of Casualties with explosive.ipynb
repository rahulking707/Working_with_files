```python
import csv
with open("Casualties with explosive1.csv","r") as file_obj:
    file_data=csv.reader(file_obj,skipinitialspace=True)
    list_file=list(file_data)
Casualties=[]
for row in list_file:
    val=row[1]
    if val!="":
        try:
            Casualties.append(int(float(val)))
        except:
            pass
            
print("total number of casualties by explosive:",sum(Casualties))
```
